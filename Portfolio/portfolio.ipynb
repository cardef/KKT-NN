```python
from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import CovarianceShrinkage
import numpy as np
import pandas as pd
import torch
```

```python
df = pd.read_csv("stock_prices.csv").drop("date", axis=1)
```

```python
sector_mapper = {
        "GOOG": "tech",
        "AAPL": "tech",
        "FB": "tech",
        "AMZN": "tech",
        "BABA": "tech",
        "GE": "utility",
        "AMD": "tech",
        "WMT": "retail",
        "BAC": "fig",
        "GM": "auto",
        "T": "auto",
        "UAA": "airline",
        "SHLD": "retail",
        "XOM": "energy",
        "RRC": "energy",
        "BBY": "retail",
        "MA": "fig",
        "PFE": "pharma",
        "JPM": "fig",
        "SBUX": "retail",
    }
```

```python
mu = mean_historical_return(df)
S = CovarianceShrinkage(df).ledoit_wolf()
```

```python
from pypfopt.efficient_frontier import EfficientFrontier
import time
sector_upper = {
        "tech": 0.2,
        "utility": 0.1,
        "retail": 0.2,
        "fig": 0.4,
        "airline": 0.05,
        "energy": 0.2}
sector_lower = {"utility": 0.01, "fig": 0.02, "airline": 0.01}

ef = EfficientFrontier(mu, S)
ef.add_sector_constraints(sector_mapper, sector_lower, sector_upper)
ef.add_constraint(lambda x: x[0] == 0.05)
ef.add_constraint(lambda x: x[0] == 0.1)
start = time.time()
ds = []
for rho in np.arange(0.001, 1.0, 0.001):
	weights = ef.max_quadratic_utility(rho)
	ds.append((2*rho - 1, [w for w in list(weights.values())]))
end = time.time()

print(end-start)
```
